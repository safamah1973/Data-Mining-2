```python
import pandas as pd

df=pd.read_csv('BostonHousing.csv')
df=df.rename(columns={'medv': 'TARGET','rm':'RM','lstat':'LSTAT'})

```

```python
df[["LSTAT","RM","TARGET"]].head()
```

```python
plt.figure(figsize=(8,4))

features =df[["LSTAT","RM"]]
target = df["TARGET"]

plt.subplot(1,2,1)
plt.scatter(df["LSTAT"],target)
plt.xlabel("LSTAT")
plt.ylabel("TARGET")

plt.subplot(1,2,2)
plt.scatter(df["LSTAT"],target)

plt.subplot(1,2,2)
plt.scatter(df["RM"],target)
plt.xlabel("RM")
plt.ylabel("TARGET")

plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(5,4))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual y")
plt.ylabel("predicted y")
plt.show()

plt.subplot(1,2,1)
```

```python
intercept=model.intercept_
coff=model.coef_
```

```python
intercept
```

```python
coff
```

```python
#Evaluate the model'sperformance
mse=mean_squared_error(y_test,y_pred)
print('Mean Square Error:',mse)
mae=mean_absolute_error(y_test,y_pred)
print('Mean absolute Error:',mae)
r2=r2_score(y_test,y_pred)
print ('R2 score:',r2)
```
